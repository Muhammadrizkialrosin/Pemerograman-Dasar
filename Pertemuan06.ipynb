```python
# SOAL 2: Menghitung Volume Kubus
def hitung_volume_kubus(sisi):
    """Menghitung volume kubus dari panjang sisinya."""
    return sisi ** 3

# Pemanggilan:
panjang_sisi = 5
print(f"Volume kubus dengan sisi {panjang_sisi} adalah: {hitung_volume_kubus(panjang_sisi)}")
```

```python
# Modifikasi: Menghitung Volume Balok
def hitung_volume_balok(panjang, lebar, tinggi):
    """Menghitung volume balok dari panjang, lebar, dan tinggi."""
    return panjang * lebar * tinggi

# Pemanggilan:
p = 10
l = 5
t = 8

volume = hitung_volume_balok(p, l, t)
print(f"Volume balok dengan panjang {p}, lebar {l}, dan tinggi {t} adalah: {volume}")
```

```python
# SOAL 3: Menghitung Rata-rata dari List Nilai
def hitung_rata_rata(data_nilai):
    """Menghitung rata-rata dari list angka menggunakan perulangan for."""
    total = 0

    # Perulangan untuk menjumlahkan nilai
    for nilai in data_nilai:
        total += nilai

    # Pembagian untuk rata-rata
    if total > 0:
        return total / len(data_nilai)
    return 0

# Pemanggilan:
nilai_praktikan = [85, 78, 92, 65, 90]
print(f"Nilai rata-rata kelas adalah: {hitung_rata_rata(nilai_praktikan)}")
```

```python
def analisis_nilai(data_nilai):
    """
    Menganalisis nilai mahasiswa dengan statistik lengkap.
    """
    total = 0
    nilai_valid = []

    # Validasi dan penjumlahan
    for nilai in data_nilai:
        if 0 <= nilai <= 100:
            nilai_valid.append(nilai)
            total += nilai

    if len(nilai_valid) == 0:
        return None

    # Hitung statistik
    rata_rata = total / len(nilai_valid)
    nilai_tertinggi = max(nilai_valid)
    nilai_terendah = min(nilai_valid)

    # Hitung jumlah mahasiswa per kategori
    lulus = sum(1 for n in nilai_valid if n >= 60)
    tidak_lulus = len(nilai_valid) - lulus

    return {
        'rata_rata': rata_rata,
        'tertinggi': nilai_tertinggi,
        'terendah': nilai_terendah,
        'total_mahasiswa': len(nilai_valid),
        'lulus': lulus,
        'tidak_lulus': tidak_lulus,
        'nilai_valid': nilai_valid
    }

# Pemanggilan:
print("ANALISIS NILAI MAHASISWA")

nilai_praktikan = [85, 78, 92, 65, 90, 55, 45, 88, 72, 95]
hasil = analisis_nilai(nilai_praktikan)

if hasil:
    print(f"\nData Nilai: {hasil['nilai_valid']}")
    print(f"\n STATISTIK:")
    print(f"   Total Mahasiswa  : {hasil['total_mahasiswa']} orang")
    print(f"   Nilai Rata-rata  : {hasil['rata_rata']:.2f}")
    print(f"   Nilai Tertinggi  : {hasil['tertinggi']}")
    print(f"   Nilai Terendah   : {hasil['terendah']}")
    print(f"\n HASIL:")
    print(f"   Lulus (≥60)      : {hasil['lulus']} orang")
    print(f"   Tidak Lulus (<60): {hasil['tidak_lulus']} orang")
    print(f"   Persentase Lulus : {(hasil['lulus']/hasil['total_mahasiswa']*100):.1f}%")
```

```python
# 4: System Penentuan Predikat Nilai

def tentukan_predikat(skor):
    """Menentukan predikat (A, B, C, D, E) berdasarkan skor."""
    if skor >= 85:
        return "A"
    elif skor >= 70:
        return "B"
    elif skor >= 60:
        return "C"
    elif skor >= 50:
        return "D"
    else:
        return "E"

# Pemanggilan:
print(f"Skor 75 mendapat predikat: {tentukan_predikat(75)}")
print(f"Skor 48 mendapat predikat: {tentukan_predikat(48)}")

```

```python
def tentukan_predikat_v2(skor):
    """Menentukan predikat (A, B, C, D, E) berdasarkan skor.
    Versi dengan validasi tipe data."""

    # Validasi tipe data - hanya menerima int atau float
    if not isinstance(skor, (int, float)):
        return "Error: Input harus berupa angka (int/float)"

    # Validasi rentang skor
    if skor < 0 or skor > 100:
        return "Error: Skor harus antara 0-100"

    # Logika penentuan predikat
    if skor >= 85:
        return "A"
    elif skor >= 70:
        return "B"
    elif skor >= 60:
        return "C"
    elif skor >= 50:
        return "D"
    else:
        return "E"

# Pemanggilan:
print(f"Skor 75 mendapat predikat: {tentukan_predikat_v2(75)}")
print(f"Skor 48 mendapat predikat: {tentukan_predikat_v2(48)}")
print(f"Skor 92.5 mendapat predikat: {tentukan_predikat_v2(92.5)}")

print("\n--- Test Validasi Tipe Data ---")
print(f"Input string '75': {tentukan_predikat_v2('75')}")
print(f"Input list [75]: {tentukan_predikat_v2([75])}")

print("\n--- Test Validasi Rentang ---")
print(f"Skor 150: {tentukan_predikat_v2(150)}")
print(f"Skor -10: {tentukan_predikat_v2(-10)}")
```

```python
print("SOAL 5: Menghitung Faktorial dengan Loop While")

def hitung_faktorial(n):
    """Menghitung faktorial dari bilangan positif n."""
    hasil = 1

    # Perulangan berjalan selama n lebih besar dari 1
    while n > 1:  # <- JAWABAN: Kondisi perulangan while
        hasil *= n
        n -= 1  # <- JAWABAN: Operasi pengurangan pada variabel n (kurangi n)

    return hasil

# Pemanggilan:
bilangan_n = 4
print(f"Faktorial dari {bilangan_n} adalah: {hitung_faktorial(bilangan_n)}")
```

```python
def hitung_faktorial_v2(n):
    """Menghitung faktorial dari bilangan positif n.
    Versi dengan validasi untuk kasus khusus."""

    # Validasi tipe data HARUS PERTAMA
    if not isinstance(n, int):
        return "Error: Input harus berupa bilangan bulat (integer)"

    # Validasi: Menangani kasus n = 0 atau n < 0
    if n == 0:
        return 1  # Faktorial dari 0 adalah 1 (definisi matematika)

    if n < 0:
        return "Error: Faktorial tidak terdefinisi untuk bilangan negatif"

    hasil = 1

    # Perulangan berjalan selama n lebih besar dari 1
    while n > 1:
        hasil *= n
        n -= 1

    return hasil

# Pemanggilan:
print("--- Test Case Normal ---")
print(f"Faktorial dari 4 adalah: {hitung_faktorial_v2(4)}")
print(f"Faktorial dari 5 adalah: {hitung_faktorial_v2(5)}")
print(f"Faktorial dari 1 adalah: {hitung_faktorial_v2(1)}")

print("\n--- Test Case Khusus ---")
print(f"Faktorial dari 0 adalah: {hitung_faktorial_v2(0)}")
print(f"Faktorial dari -5 adalah: {hitung_faktorial_v2(-5)}")

print("\n--- Test Validasi Tipe Data ---")
print(f"Input float 4.5: {hitung_faktorial_v2(4.5)}")
print(f"Input string '4': {hitung_faktorial_v2('4')}")
```

```python
print("SOAL 6: Analisis Data Suhu Ekstrem (for dan if/elif)")

def analisis_suhu(data_suhu):
    """Menghitung jumlah hari Panas dan Dingin dari data_suhu."""
    panas_count = 0
    dingin_count = 0  # <- JAWABAN 1: Inisialisasi dingin_count

    for suhu in data_suhu:
        if suhu >= 35:
            panas_count += 1
        elif suhu <= 15:  # <- JAWABAN 2: Kondisi untuk hari dingin
            dingin_count += 1

    return panas_count, dingin_count

# Pemanggilan:
suhu_mingguan = [36, 30, 14, 25, 38, 16, 28]
jumlah_panas, jumlah_dingin = analisis_suhu(suhu_mingguan)
print(f"Jumlah hari Panas (>=35°C): {jumlah_panas}")
print(f"Jumlah hari Dingin (<=15°C): {jumlah_dingin}")
```

```python
def analisis_suhu_v2(data_suhu, batas_panas=35, batas_dingin=15):
    """Menghitung jumlah hari Panas dan Dingin dari data_suhu.
    Versi dengan parameter opsional untuk batas suhu."""

    panas_count = 0
    dingin_count = 0

    for suhu in data_suhu:
        if suhu >= batas_panas:
            panas_count += 1
        elif suhu <= batas_dingin:
            dingin_count += 1

    return panas_count, dingin_count

# Pemanggilan dengan nilai default:
print("--- Test dengan Nilai Default (35 dan 15) ---")
suhu_mingguan = [36, 30, 14, 25, 38, 16, 28]
jumlah_panas, jumlah_dingin = analisis_suhu_v2(suhu_mingguan)
print(f"Jumlah hari Panas (>=35°C): {jumlah_panas}")
print(f"Jumlah hari Dingin (<=15°C): {jumlah_dingin}")

# Pemanggilan dengan parameter kustom:
print("\n--- Test dengan Batas Kustom (30 dan 20) ---")
jumlah_panas, jumlah_dingin = analisis_suhu_v2(suhu_mingguan, batas_panas=30, batas_dingin=20)
print(f"Jumlah hari Panas (>=30°C): {jumlah_panas}")
print(f"Jumlah hari Dingin (<=20°C): {jumlah_dingin}")

# Test case tambahan:
print("\n--- Test dengan Data Berbeda ---")
suhu_bulan = [40, 12, 35, 28, 15, 8, 37, 22, 10, 33]
jumlah_panas, jumlah_dingin = analisis_suhu_v2(suhu_bulan)
print(f"Data: {suhu_bulan}")
print(f"Jumlah hari Panas (>=35°C): {jumlah_panas}")
print(f"Jumlah hari Dingin (<=15°C): {jumlah_dingin}")

print("\n--- Penjelasan Proses--- ")
print("Data: [36, 30, 14, 25, 38, 16, 28]")
print("36°C  → Panas (>=35)")
print("30°C  → Normal")
print("14°C  → Dingin (<=15)")
print("25°C  → Normal")
print("38°C  → Panas (>=35)")
print("16°C  → Normal")
print("28°C  → Normal")
print("Hasil: 2 hari Panas, 1 hari Dingin")
```
