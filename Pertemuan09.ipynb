```python
import pandas as pd

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

city_names = ['Jakarta', 'Bandung', 'Makassar', 'Surabaya', 'Medan', 'Yogyakarta', 'Malang']
population = [498044, 964254, 491918, 8398748, 653115, 883305, 744955]
num_airports = [2, 2, 8, 3, 1, 3, 2]

df = pd.DataFrame({
    'City Name': city_names,
    'Population': population,
    'Airports': num_airports,
})

print("=" * 60)
print("DATASET KOTA-KOTA DI INDONESIA")
print("=" * 60)
print(df)
print("\n")

print("=" * 60)
print("1. RINGKASAN STATISTIK LENGKAP")
print("=" * 60)
print(df.describe())
print("\n")

print("PENJELASAN describe():")
print("- count: Jumlah data")
print("- mean: Rata-rata")
print("- std: Standar deviasi")
print("- min: Nilai minimum")
print("- 25%: Kuartil pertama (Q1)")
print("- 50%: Median (Q2)")
print("- 75%: Kuartil ketiga (Q3)")
print("- max: Nilai maksimum")
print("\n")

print("=" * 60)
print("2. RATA-RATA POPULASI")
print("=" * 60)

rata_rata_populasi = df['Population'].mean()
print(f"Rata-rata populasi: {rata_rata_populasi:,.2f} jiwa")

print("\nINTERPRETASI:")
print(f"Rata-rata populasi dari {len(df)} kota adalah {rata_rata_populasi:,.0f} jiwa.")
print("\nKOTA DENGAN POPULASI DI ATAS RATA-RATA:")
kota_atas_rata = df[df['Population'] > rata_rata_populasi]
for idx, row in kota_atas_rata.iterrows():
    print(f"- {row['City Name']}: {row['Population']:,} jiwa")
print("\n")

plt.figure(figsize=(12, 6))
colors = ['red' if x > rata_rata_populasi else 'skyblue' for x in df['Population']]
bars = plt.bar(df['City Name'], df['Population'], color=colors, edgecolor='black', linewidth=1.2)
plt.axhline(y=rata_rata_populasi, color='green', linestyle='--', linewidth=2, label=f'Rata-rata: {rata_rata_populasi:,.0f}')
plt.xlabel('Kota', fontsize=12, fontweight='bold')
plt.ylabel('Populasi', fontsize=12, fontweight='bold')
plt.title('Populasi Kota-kota di Indonesia', fontsize=14, fontweight='bold', pad=20)
plt.xticks(rotation=45, ha='right')
plt.legend(fontsize=10)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("=" * 60)
print("3. MODUS JUMLAH BANDARA")
print("=" * 60)

modus_airports = df['Airports'].mode()[0]
print(f"Modus jumlah Airports: {modus_airports}")

print("\nDISTRIBUSI FREKUENSI AIRPORTS:")
frekuensi = df['Airports'].value_counts().sort_index()
print(frekuensi)

print(f"\nINTERPRETASI:")
print(f"Modus = {modus_airports} artinya sebagian besar kota memiliki {modus_airports} bandara.")
print(f"Jumlah kota dengan {modus_airports} bandara: {frekuensi[modus_airports]} kota")
print("\n")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

frekuensi.plot(kind='bar', ax=axes[0], color='coral', edgecolor='black', linewidth=1.5)
axes[0].set_xlabel('Jumlah Bandara', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Jumlah Kota', fontsize=12, fontweight='bold')
axes[0].set_title('Distribusi Jumlah Bandara per Kota', fontsize=13, fontweight='bold', pad=15)
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=0)
axes[0].grid(axis='y', alpha=0.3)

colors_pie = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
axes[1].pie(frekuensi, labels=[f'{int(x)} bandara' for x in frekuensi.index],
            autopct='%1.1f%%', startangle=90, colors=colors_pie[:len(frekuensi)],
            wedgeprops={'edgecolor': 'black', 'linewidth': 1.5})
axes[1].set_title(f'Persentase Kota Berdasarkan Jumlah Bandara\n(Modus = {modus_airports})',
                  fontsize=13, fontweight='bold')

plt.tight_layout()
plt.show()


print("=" * 60)
print("4. STATISTIK TAMBAHAN")
print("=" * 60)

median_pop = df['Population'].median()
print(f"Median populasi: {median_pop:,.2f} jiwa")

std_pop = df['Population'].std()
print(f"Standar deviasi populasi: {std_pop:,.2f}")

range_pop = df['Population'].max() - df['Population'].min()
print(f"Range populasi: {range_pop:,} jiwa")

var_pop = df['Population'].var()
print(f"Variance populasi: {var_pop:,.2f}")

print("\n")

plt.figure(figsize=(10, 6))
box = plt.boxplot(df['Population'], vert=True, patch_artist=True,
                   boxprops=dict(facecolor='lightblue', edgecolor='black', linewidth=1.5),
                   medianprops=dict(color='red', linewidth=2),
                   whiskerprops=dict(color='black', linewidth=1.5),
                   capprops=dict(color='black', linewidth=1.5))
plt.title('Boxplot Distribusi Populasi', fontsize=14, fontweight='bold', pad=15)
plt.ylabel('Populasi', fontsize=12, fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("=" * 60)
print("5. INFORMASI KOTA TERBESAR DAN TERKECIL")
print("=" * 60)

kota_terbesar = df.loc[df['Population'].idxmax()]
print(f"Kota terbesar: {kota_terbesar['City Name']}")
print(f"Populasi: {kota_terbesar['Population']:,} jiwa")
print(f"Airports: {kota_terbesar['Airports']}")

print()

kota_terkecil = df.loc[df['Population'].idxmin()]
print(f"Kota terkecil: {kota_terkecil['City Name']}")
print(f"Populasi: {kota_terkecil['Population']:,} jiwa")
print(f"Airports: {kota_terkecil['Airports']}")

print("\n")

plt.figure(figsize=(10, 6))
plt.scatter(df['Airports'], df['Population'], s=200, c='steelblue',
            edgecolors='black', linewidth=1.5, alpha=0.7)

for i, row in df.iterrows():
    plt.annotate(row['City Name'],
                (row['Airports'], row['Population']),
                xytext=(5, 5), textcoords='offset points',
                fontsize=9, fontweight='bold')

plt.xlabel('Jumlah Bandara', fontsize=12, fontweight='bold')
plt.ylabel('Populasi', fontsize=12, fontweight='bold')
plt.title('Hubungan Populasi dan Jumlah Bandara', fontsize=14, fontweight='bold', pad=15)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("=" * 60)
print("JAWABAN TUGAS")
print("=" * 60)

print(f"\n1. Kota mana yang populasinya mendekati rata-rata?")
selisih = abs(df['Population'] - rata_rata_populasi)
kota_terdekat = df.loc[selisih.idxmin()]
print(f"   Jawaban: {kota_terdekat['City Name']}")
print(f"   Populasi: {kota_terdekat['Population']:,} jiwa")
print(f"   Selisih dari rata-rata: {selisih.min():,.0f} jiwa")
print(f"   Persentase selisih: {(selisih.min()/rata_rata_populasi)*100:.2f}%")

print(f"\n2. Apa arti modus untuk kolom Airports?")
print(f"   Jawaban: Modus = {modus_airports}")
print(f"   Artinya: Nilai yang paling sering muncul adalah {modus_airports}")
print(f"   Sebagian besar kota ({frekuensi[modus_airports]} dari {len(df)} kota) memiliki {modus_airports} bandara.")
if modus_airports == 0:
    print(f"   Ini menunjukkan mayoritas kota tidak memiliki bandara.")
elif modus_airports == 1:
    print(f"   Ini menunjukkan mayoritas kota memiliki 1 bandara.")
else:
    print(f"   Ini menunjukkan mayoritas kota memiliki {modus_airports} bandara.")

print("\n" + "=" * 60)
print("SELESAI - PRAKTIKUM STATISTIK DESKRIPTIF")
print("=" * 60)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

city_names = ['Jakarta', 'Bandung', 'Makassar', 'Surabaya', 'Medan', 'Yogyakarta', 'Malang']
population = [498044, 964254, 491918, 8398748, 653115, 883305, 744955]
num_airports = [2, 2, 8, 3, 1, 3, 2]

df = pd.DataFrame({
    'City Name': city_names,
    'Population': population,
    'Airports': num_airports,
})

print("=" * 70)
print("DATASET KOTA-KOTA DI INDONESIA")
print("=" * 70)
print(df)
print("\n")

print("=" * 70)
print("HISTOGRAM - DISTRIBUSI POPULASI KOTA")
print("=" * 70)

plt.figure(figsize=(12, 6))
df['Population'].hist(bins=5, color='skyblue', edgecolor='black', linewidth=1.5, alpha=0.7)
plt.title("Distribusi Populasi Kota", fontsize=16, fontweight='bold', pad=20)
plt.xlabel("Jumlah Penduduk", fontsize=13, fontweight='bold')
plt.ylabel("Frekuensi", fontsize=13, fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("\nSTATISTIK POPULASI:")
print(f"Rata-rata    : {df['Population'].mean():,.2f} jiwa")
print(f"Median       : {df['Population'].median():,.2f} jiwa")
print(f"Std Deviasi  : {df['Population'].std():,.2f}")
print(f"Min          : {df['Population'].min():,} jiwa ({df.loc[df['Population'].idxmin(), 'City Name']})")
print(f"Max          : {df['Population'].max():,} jiwa ({df.loc[df['Population'].idxmax(), 'City Name']})")
print(f"Range        : {df['Population'].max() - df['Population'].min():,} jiwa")

print("\n")

print("=" * 70)
print("TUGAS: PENGAMATAN DARI HISTOGRAM")
print("=" * 70)

mean_pop = df['Population'].mean()
median_pop = df['Population'].median()
max_pop = df['Population'].max()
min_pop = df['Population'].min()

print("\n1. BENTUK DISTRIBUSI:")
if mean_pop > median_pop * 1.5:
    print("   - Distribusi sangat SKEWED RIGHT (miring kanan/positif)")
    print("   - Ada outlier yang sangat besar (Surabaya)")
elif mean_pop > median_pop:
    print("   - Distribusi SKEWED RIGHT (miring kanan)")
    print("   - Beberapa kota memiliki populasi yang jauh lebih tinggi")
else:
    print("   - Distribusi relatif SIMETRIS")

print(f"\n2. KONSENTRASI DATA:")
# Hitung berapa kota di setiap rentang
rentang_bawah = df[df['Population'] < 1000000]
rentang_atas = df[df['Population'] >= 1000000]
print(f"   - Kota dengan populasi < 1 juta: {len(rentang_bawah)} kota ({len(rentang_bawah)/len(df)*100:.1f}%)")
print(f"   - Kota dengan populasi  1 juta: {len(rentang_atas)} kota ({len(rentang_atas)/len(df)*100:.1f}%)")

print(f"\n3. OUTLIER:")
if max_pop > mean_pop * 3:
    print(f"   - Surabaya adalah OUTLIER yang signifikan")
    print(f"   - Populasinya ({max_pop:,}) adalah {max_pop/mean_pop:.1f}x dari rata-rata")
    print(f"   - Ini membuat distribusi sangat tidak merata")

print(f"\n4. VARIABILITAS:")
cv = (df['Population'].std() / df['Population'].mean()) * 100
print(f"   - Coefficient of Variation: {cv:.2f}%")
if cv > 50:
    print(f"   - Variabilitas SANGAT TINGGI (CV > 50%)")
    print(f"   - Data populasi sangat bervariasi antar kota")
elif cv > 30:
    print(f"   - Variabilitas TINGGI (CV > 30%)")
else:
    print(f"   - Variabilitas SEDANG")

print("\n5. INTERPRETASI HISTOGRAM:")
print("   Dari histogram dapat dilihat bahwa:")
print("   - Mayoritas kota memiliki populasi di rentang rendah (< 1 juta)")
print("   - Ada satu kota (Surabaya) dengan populasi yang jauh lebih tinggi")
print("   - Distribusi tidak normal/merata karena adanya outlier")
print("   - Perbedaan populasi antar kota sangat besar")

print("\n" + "=" * 70)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

axes[0, 0].hist(df['Population'], bins=3, color='lightcoral', edgecolor='black', linewidth=1.5)
axes[0, 0].set_title('Histogram (3 bins)', fontsize=12, fontweight='bold')
axes[0, 0].set_xlabel('Populasi', fontweight='bold')
axes[0, 0].set_ylabel('Frekuensi', fontweight='bold')
axes[0, 0].grid(axis='y', alpha=0.3)

axes[0, 1].hist(df['Population'], bins=7, color='lightgreen', edgecolor='black', linewidth=1.5)
axes[0, 1].set_title('Histogram (7 bins)', fontsize=12, fontweight='bold')
axes[0, 1].set_xlabel('Populasi', fontweight='bold')
axes[0, 1].set_ylabel('Frekuensi', fontweight='bold')
axes[0, 1].grid(axis='y', alpha=0.3)

box = axes[1, 0].boxplot(df['Population'], vert=True, patch_artist=True,
                         boxprops=dict(facecolor='lightblue', edgecolor='black', linewidth=1.5),
                         medianprops=dict(color='red', linewidth=2))
axes[1, 0].set_title('Boxplot Distribusi', fontsize=12, fontweight='bold')
axes[1, 0].set_ylabel('Populasi', fontweight='bold')
axes[1, 0].grid(axis='y', alpha=0.3)

axes[1, 1].bar(df['City Name'], df['Population'], color='steelblue', edgecolor='black', linewidth=1.2)
axes[1, 1].set_title('Populasi per Kota', fontsize=12, fontweight='bold')
axes[1, 1].set_xlabel('Kota', fontweight='bold')
axes[1, 1].set_ylabel('Populasi', fontweight='bold')
axes[1, 1].tick_params(axis='x', rotation=45)
axes[1, 1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("\nVISUALISASI SELESAI!")
print("Perhatikan bagaimana jumlah bins mempengaruhi tampilan histogram.")
print("=" * 70)
```
